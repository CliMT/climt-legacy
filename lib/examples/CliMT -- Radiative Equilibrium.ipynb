# Radiative Equilibrium
----------------------------------

This demo steps the grey radiation code forward in time to get a radiative equilibrium
profile. The surface temperature is 283 K, and the maximum atmospheric temperature is
around 255 K, making this profile highly unstable without a boundary layer or convection.

```python
%matplotlib notebook
import numpy as np
import climt

# Federation
kwargs = {}

kwargs['MonitorFields'] = ['T']
kwargs['solin'] = 400.
rad = climt.radiation(scheme='newgreygas', **kwargs)

kwargs['q'] = np.zeros(rad.nlev) + 1.e-9
kwargs['T'] = np.zeros(rad.nlev) + (kwargs['solin']/2./5.67e-8)**0.25


# -- Initial fields (temp is set to skin temp)

rad['T'] = kwargs['T']

print rad['T']

for i in range(100000):
    rad.step()
```
